```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
from typing import List, Tuple
```

```python
with open('inputs/dec05.txt', 'r') as f:
    lines = f.readlines()
lines = [line.strip() for line in lines]
rules = [[int(s) for s in l.split("|")] for l in lines if '|' in l]
updates = [[int(s) for s in l.split(",")] for l in lines if ',' in l]
updates
```

```python
def is_rule_applicable(update: List[str], rule: List[str]) -> bool:
    X = rule[0]
    Y = rule[1]
    if X in update and Y in update:
        return True
    return False

def is_rule_satisfied(update: List[str], rule: List[str]) -> bool:
    X = rule[0]
    Y = rule[1]
    
    iX = update.index(X)
    iY = update.index(Y)
    # print(X, Y, iX, iY)
    if iX < iY:
        return True
    return False

def is_valid_update(update: List[str], rules: List[str]) -> bool:
    count_applicable = 0
    count_satisfied = 0
    for rule in rules:
        if not is_rule_applicable(update, rule):
            continue
        count_applicable += 1
        if is_rule_satisfied(update, rule):
            count_satisfied += 1
    if count_applicable == count_satisfied:
        return True
    return False

def check_updates(updates: List[List[str]], rules: List[List[str]]) -> List[bool]:
    return [is_valid_update(update, rules) for update in updates]

def get_middle_number(update: List[str]) -> int:
    return update[len(update) // 2]
```

```python
def get_score(updates: List[List[str]], rules: List[List[str]]) -> int:
    is_valid = check_updates(updates, rules)
    valid_updates = [u for iu, u in enumerate(updates) if is_valid[iu]]
    print(valid_updates)
    return sum([get_middle_number(u) for u in valid_updates])

get_score(updates, rules)
```

```python
def fix_update(update: List[int], rules: List[List[int]]) -> List[int]:
    for rule in rules:
        X = rule[0]
        Y = rule[1]
        if X in update and Y in update:
            iX = update.index(X)
            iY = update.index(Y)
            if iX > iY:
                update[iX], update[iY] = update[iY], update[iX]
    return update

def get_score_fixed(updates: List[List[int]], rules: List[List[int]]) -> int:
    is_valid = check_updates(updates, rules)
    invalid_updates = [u for iu, u in enumerate(updates) if not is_valid[iu]]
    while True:
        is_invalid_valid = check_updates(invalid_updates, rules)
        if all(is_invalid_valid):
            break
        fixed_updates = [fix_update(u, rules) for u in invalid_updates]
        invalid_updates = fixed_updates
    print(fixed_updates)
    return sum([get_middle_number(u) for u in fixed_updates])

get_score_fixed(updates, rules)
```
