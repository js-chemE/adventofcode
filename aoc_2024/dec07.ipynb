```python
def int2bool(integer: int) -> List[bool]:
    binary_str = bin(integer)[2:]  # Remove the '0b' prefix
    return [bool(int(b)) for b in binary_str]

def array2operators(array: List[bool]) -> List[str]:
    return ['+' if b else '*' for b in array]

def int2operators(integer: int) -> List[str]:
    return array2operators(int2bool(integer))
```

```python
OPERATERS = ["+", "*"]

def operate1(number0: int, number1: int, operator: str) -> int:
    if operator == "+":
        return number0 + number1
    elif operator == "-":
        return number0 - number1
    elif operator == "*":
        return number0 * number1
    elif operator == "/":
        return number0 / number1
    else:
        raise ValueError("Invalid operator")

def evaluate1_single(solution: int, numbers: List[int], operators: List[str]) -> bool:
    if len(numbers) != len(operators) + 1:
        raise ValueError(f"Invalid input. Number of numbers ({len(numbers)}) must be one more than number of operators ({len(operators)})")
    result = numbers[0]
    #print(result)
    for i in range(len(operators)):
        result = operate1(result, numbers[i + 1], operators[i])
        #print(operators[i], numbers[i + 1], result)
        if result > solution:
            return False
    if result == solution:
        return True
    else:
        return False


def evaluate1(solution=List[int], numbers=List[int]) -> bool:
    operators = itertools.product(OPERATERS, repeat=len(numbers) - 1)
    for ops in operators:
        #print(ops)
        if evaluate1_single(solution, numbers, ops):
            return True
    return False

def evaluate1_values(solution: int, numbers: List[int]) -> int:
    if evaluate1(solution, numbers):
        return solution
    return 0

def test1(solutions: List[int], set_numbers: List[List[int]]) -> List[bool]:
    return np.array([evaluate1_values(solution, numbers) for solution, numbers in zip(solutions, set_numbers)])

def count1(solutions: List[int], set_numbers: List[List[int]]) -> int:
    return test1(solutions, set_numbers).sum()
        

i_test = 8
print(solutions[i_test], set_numbers[i_test])
#evaluate1(solutions[i_test], set_numbers[i_test])
count1(solutions, set_numbers)
```
