```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
reports = []
with open('dec02.txt', 'r') as file:
    for line in file:
        reports.append(np.array([int(n) for n in line.strip().split(" ")]))  # .strip() removes leading/trailing whitespace
print(reports)
```

```python
def check_report(report: np.ndarray) -> bool:
    steps = np.diff(report)
    return (np.all(steps > 0) or np.all(steps < 0)) and (np.all(np.abs(steps) >= 1) and np.all(np.abs(steps) <= 3))

def check_unsafe_report(report: np.ndarray) -> bool:
    is_safe = False
    for ilevel, level in enumerate(report):
        if check_report(np.delete(report, ilevel)):
            is_safe = True
            return is_safe
    return is_safe
```

```python
masks = np.zeros(len(reports), dtype=bool)
for i, report in enumerate(reports):
    masks[i] = check_report(report)
number_safe_reports = np.sum(masks)
print(number_safe_reports)
```

```python
false_reports = [r for ir, r in enumerate(reports) if not masks[ir]]
safe_unsafe_mask = np.zeros(len(false_reports), dtype=bool)
for i, report in enumerate(false_reports):
    safe_unsafe_mask[i] = check_unsafe_report(report)
    print(safe_unsafe_mask[i], report)
number_safe_unsafe_reports = np.sum(safe_unsafe_mask)
print(number_safe_unsafe_reports)
```
